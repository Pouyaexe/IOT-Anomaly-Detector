```python
import os
import subprocess

# Set the path to the dataset folder
dataset_folder = "dataset_folder"

# Function to unrar files
def unrar_files(directory):
    directory_path = os.path.join(dataset_folder, directory)
    for file in os.listdir(directory_path):
        if file.endswith(".rar"):
            rar_file_path = os.path.join(directory_path, file)
            extraction_dir = os.path.join(directory_path, file.split(".")[0])
            os.makedirs(extraction_dir, exist_ok=True)
            subprocess.call(["unrar", "x", rar_file_path, extraction_dir])

# Iterate through the directories and unrar the files
for directory in os.listdir(dataset_folder):
    directory_path = os.path.join(dataset_folder, directory)
    if os.path.isdir(directory_path):
        unrar_files(directory)

```
